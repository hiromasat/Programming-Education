## 2. じゃんけんゲームを作ろう

### シンプルなじゃんけん

```python
import random

# じゃんけんのリスト
janken_hand = ["グーです。", "チョキです。", "パーです。"]

print("じゃんけんAIと「じゃんけん」をしよう！")
print("\n0：グー\n1:チョキ\n2:パー")

# じゃんけんの入力
my_janken = int(input("じゃんけんの手を決めてください。数値を入力後「enter」キーを押してね。："))

# パソコンの手
janken_pc = random.randint(0, 2)

print("出した手：", janken_hand[my_janken])

print("じゃんけんPC", janken_hand[janken_pc])

hantei =(my_janken - janken_pc + 3) % 3

if hantei == 0:
  print("あいこです。")
elif hantei == 1:
  print("あなたの負けです。")
else:
  print("あなたの勝ちです。")
```

### 2. ちょっとこだわったじゃんけん

```python
import random

janken_hand = ["グー", "チョキ", "パー"]

# 初期化
result = 0

# じゃんけんのスコアを記録
score = 0

# じゃんけんの回数
trial = 5
```

```python
print("「じゃんけん」をしよう！「じゃんけん、、、、」")
print("(0：グー 1:チョキ 2:パー)")
print()

# じゃんけんゲーム
def janken_game():

  # 自分の手を決める
  my_janken = what_my_janken()

  # 0~2の数字をランダムに得る
  janken = random.randint(0, 2)
  result =(my_janken - janken + 3) % 3

  print("出した手：", janken_hand[my_janken])
  print("じゃんけんくん:", janken_hand[janken])

  return result


# 自分の手を決める
def what_my_janken() -> int:
    my_janken = input("じゃんけんの手を決めてね。\n")
    try:
        my_janken = int(my_janken)
    except:
        print("0, 1, 2の数字を入力してください")
        return what_my_janken()
    if my_janken < 0 or my_janken > 2:
        print("0, 1, 2の数字を入力してください")
        return what_my_janken()
    else:
        return my_janken


# 判定
def judgement(result):
    if result == 0:
      print('\nあいこで')
      return 0

    elif result == 1:
      print('\n君の負け')
      return 0
      
    else :
      print('\n君の勝ち')
      return 1


for i in range(trial):
  print(f'{i+1}回目')

  # 勝負がつくまで繰り返す
  while not result:
    result = janken_game()
    point = judgement(result)
    
  # scoreにポイントを加算する
  score += point
  
  # pointとresultを0に戻す
  point, result = 0, 0 
```

```python
from google.colab.patches import cv2_imshow
import cv2

def score_publication(score):
  if score > trial // 2:
    print('優勝')
    img = cv2.imread('trophy_boy.jpg')

  else:
    print('また挑戦してね')
    img = cv2.imread('kuyashii.jpg')
  return img

print(f'あなたのスコアは{score}点です。')
img = score_publication(score)
# 画像の表示
cv2_imshow(img)
```
